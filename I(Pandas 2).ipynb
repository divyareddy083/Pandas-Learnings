```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'], 'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],'value': [5, 6, 7, 8]})
df1.merge(df2, left_on='lkey', right_on='rkey')
```

```python
np.random.seed(20)
df1=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('2022-04-29',periods=10),columns=['A','B','C','D'])
df1
```

```python
df1['New']=df1['C']+df1['D']
df1
```

```python
df1['New_col_sqrtD']=df1['D'].apply(lambda x : x**2)
df1
```

```python
df1['D_log']=df1['D'].apply(lambda x : np.log(x))
df1 #-ve values leads to nan values
```

```python
def test(x):
    if (x<0):
        return x
    else:
        return np.log(x)
```

```python
df1['New_D_log']=df1['D'].apply(test)
```

```python
df1
```

```python
def test1(x):
    if x<0:
        return 1
    else:
        return 0
```
