#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# View the plot 
%matplotlib inline
```

```python
#cause of deaths
cause_of_deaths=['Chronic Disease','Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis','Others']
#percentile
percentiles=[62, 5, 4, 2, 1 , 26 ]

df=pd.DataFrame([cause_of_deaths,percentiles]).T
df['cause_of_deaths']=df[0]
df['percentiles']=df[1]
df.drop([0,1],axis=1,inplace=True)
df
```

#### 2: Draw a pie chart using the two data points

```python
#set the pie chart plot properties
#set figure size
plt.figure(figsize=(10,10))
#explode the largest pie in the data set
explode=[0.05,0,0,0,0,0]
plt.pie(percentiles,labels=cause_of_deaths,explode=explode,autopct='%1.1f%%',startangle=70)
#set axis equal to draw pie as circle
plt.axis('equal')
#set title of the pie chart
plt.title('ohio state data')
#show the plot (pie chart)
plt.show()
```
