# Seaborn Library

## Introduction

>Seaborn is a data visualization library built on top of core visualization library Matplotlib.It is considered as a complement rather than a replacement for Matplotlib library.It also closely integrated to the data structures from pandas.It provides high-level interface for drawing attractive statistical graphics.

### Different categories of plot in Seaborn 

>Plots are basically used for visualizing the relationship between variables. Those variables can be either be completely numerical or a category like a group, class or division. Seaborn divides plot into the below categories : 
>><ul><li>Relational plots: This plot is used to understand the relation between two variables.</li>
>><li>Categorical plots: This plot deals with categorical variables and how they can be visualized.</li>
>><li>Distribution plots: This plot is used for examining univariate and bivariate distributions.</li>
>><li>Regression plots: The regression plots in seaborn are primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses.</li>
>><li>Matrix plots: A matrix plot is an array of scatterplots.</li>
>><li>Multi-plot grids: It is an useful approach is to draw multiple instances of the same plot on different subsets of the dataset.</li></ul>

# Box Plot

>A box plot describes the **distribution of a continous variable by plotting the summary of five statistical terms:minimum,maximum,first quartile(25<sup>th</sup>percentile),second quartile(50<sup>th</sup>percentile),and third quartile(75<sup>th</sup>percentile)** in the dataset.Box plots are measure of how well the data is distributed in the dataset.
>>A box plot helps to maintain the **distribution of quantitative data** in such a way that it facilitates the comparisons between variables or across levels of a **categorical variable**. 
>>>It is also possible to draw multiple boxes containing a categorical variable with different values.It is also useful in comparing the distribution of **data across dataset by drawing box plots for each of them**.
>>>>Box plots have **vertical lines** extending from the boxes which are termed as **whiskers**.These whiskers indicates variability outside the upper and lower quartiles,hence the boxplots are also known as **box-and whisker plots/diagrams**.
>>>>>Any **outlier** in the dataset are plotted as **individual points**.
>>>>>> Boxplots are created using **boxplot()**.

## Visual Explanation of Boxplot

<img src="https://miro.medium.com/max/1400/1*2c21SkzJMf3frPXPAR_gZA.png" alt="Box Plot" height=600px width=600px>

<ul><li>median (Q2/50th Percentile): the middle value of the dataset.</li>
<li>first quartile (Q1/25th Percentile): the middle number between the smallest number (not the “minimum”) and the median of the dataset.</li>
<li>third quartile (Q3/75th Percentile): the middle value between the median and the highest value (not the “maximum”) of the dataset.</li>
<li>interquartile range (IQR): 25th to the 75th percentile.IQR=Q3-Q1 </li>
<li>whiskers (shown in blue)</li>
<li>outliers (shown as green circles)</li>
<li>“maximum”: Q3+1.5*IQR</li>
<li>“minimum”: Q1-1.5*IQR</li></ul>

# Violin plot
>A violin plot is a combination of **box plot and kernel density plot**.Hence,violin plot makes it easier to analyse and understand the distribution of the data,because it includes both types of plots.
>>A violin plot is a **hybrid of a box plot and a kernel density plot**, which shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.
>>>Similar to box plot,a violin plot can produce **multiple violins** corresponding to different values of a **categorical variable**
>>>>The **quartile** and **whisker** values from the boxplot are shown inside the violin.As the violin plot uses KDE,the **wider portion** of violin indicates **higher density** and the **narrow region** reprsents relatively **lower density**.
The **inter-quartile range** in boxplot and higher density portion in "KDE" fall in the same region of each category of violin plot.

## Visual Explaination of Violin Plot
<img src="https://diagrammm.com/img/diagrams/violin-plot-terminology.svg" alt="Violin Plot" height=600px width=600px>

>As a Violin Plot is an extended version of the **Box Plot** it has many of the same summary statistics. For instance, the white dot represents the **median**. The black bar in the middle of the spot represents the **interquartile range** and the vertical thin line - **95% confidence interval**. On each side of the vertical line is a **kernel density estimation (Density Plot)** that shows distribution shape of the data. **Wider** sections represent **higher probability**, the **skinnier - lower probability**.
>>By comparing Violine Plor and the plain Boxplot it is easy to notice that the **Violine Plot is less popular**. The reason is that sometimes it might be difficult to read for those who are not familiar with such type of the chart.

```python
# Creating violin plot for "age" w.r.t "sex"
sns.violinplot(x='class',y='age',hue='sex',data=titanic).set_title("Violin plot")
```

# Factor Plot

## Visual Explaination of Correlation 

##### What is correlation?
>Correlation is a statistical measure that expresses the extent to which two variables are linearly related (meaning they change together at a constant rate). It’s a common tool for describing simple relationships without making a statement about cause and effect.

##### How is correlation measured?
>The sample correlation coefficient, r, quantifies the strength of the relationship. Correlations are also tested for statistical significance.

##### What are some limitations of correlation analysis?
>Correlation can’t look at the presence or effect of other variables outside of the two being explored. Importantly, correlation doesn’t tell us about cause and effect. Correlation also cannot accurately describe curvilinear relationships.

##### What do correlation numbers mean?
>We describe correlations with a unit-free measure called the correlation coefficient which ranges from -1 to +1 and is denoted by r. Statistical significance is indicated with a p-value. Therefore, correlations are typically written with two key numbers: r = and p = .
>><ul><li>The closer r is to zero, the weaker the linear relationship.</li>
>><li>Positive r values indicate a positive correlation, where the values of both variables tend to increase together.</li>
>><li>Negative r values indicate a negative correlation, where the values of one variable tend to increase when the values of the other variable decrease.</li>
>><li>The p-value gives us evidence that we can meaningfully conclude that the population correlation coefficient is likely different from zero, based on what we observe from the sample.</li>
>><li>"Unit-free measure" means that correlations exist on their own scale</li></ul>

##### What is a p-value?
>A p-value is a measure of probability used for hypothesis testing

# Univariate Distribution Plot

>Distribution of data is the foremost thing that we need to understand while analysing the data.The **displot()** function helps us to visualize the **parametric distribution of dataset**.
>>We can understand the **univariate distribution** of the data by plotting a histogram that fits the kernel density estimation od data.The 2 most important arguments are **kde(Kernel density plot) and hist(histogram)** which has a **Boolean value**.If the values are **True** the plot shows **both the charts**.
>>> KDE is aa way to estimate the **probability density function** of a **continous variable**.
>>>>The **bins** is an important argument which defined the **number of bins in the agument**.

```python
plt.figure(figsize=(12,11))

# Univariate distribution plot-histogram for "petal_length"
plt.subplot(311)
sns.distplot(iris[['petal_length']]).set_title("Distribution Plot ")

# Univariate distribution plot-histogram for "sepal_length"
plt.subplot(312)
sns.distplot(iris[['petal_length']],kde=False).set_title("Distribution Plot with kde=False")

# Univariate distribution plot-histogram for "sepal_length"
plt.subplot(313)
sns.distplot(iris[['petal_length']],hist=False).set_title("Distribution Plot with hist=False")

plt.show()
```

### set_palette()
> Used to give colors to plots and adding more aesthitic values to it.we can basically create a **colored sequential plot or diverging plot**.
>>**Sequential plots** are suitable to express the distribution of  data ranging from relative lower values to higher values within a range .we need to add characters to the color passed to plot the sequential plot

```python
#importing libraries to create a function

import numpy as np
from matplotlib import pyplot as plt
def mychart(num=1):
    value=np.linspace(1,10,100)
    for i in range(1,10):
        plt.plot(value,np.sin(value+i*8)*(10-i)*num)
sns.set_style("white")

#creating a figure of size
plt.figure(1)
plt.figure(figsize=(15,8))

#adjusting the settings to sequential plot to red color
sns.set_palette("Reds")
plt.subplot(121)
mychart()

#adjusting the settings to diverging palette from brown color
sns.set_palette("BrBG",10)
plt.subplot(122)
mychart()

plt.show()
```

The first chart above is created used default red color ,second chart is created diverging from brown to blue green color
