#### Matplotlib Labels and Title
##### Labels for a Plot
>With Pyplot, we can use the **xlabel() and ylabel()** functions to set a label for the x- and y-axis
##### Title for a Plot
>With Pyplot, you can use the **title()** function to set a title for the plot.

```python
x = np.array([20,80, 85, 90, 95, 100, 105, 110, 115, 120, 125,128])
y = np.array([210,240, 250, 260, 270, 280, 290, 300, 310, 320, 330,355])
plt.plot(x, y,lw=1.5,c='purple',ls='-',marker='o',ms=8,mec='purple',mfc='r')
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

##### Set Font Properties for Title and Labels
>You can use the **fontdict parameter in xlabel(), ylabel(), and title()** to set font properties for the title and labels.

##### Position the Title
> You can use the **loc parameter in title() to position the title**.
> Legal values are: **'left', 'right', and 'center'. Default value is 'center'**.

```python
x = np.array([20,80, 85, 90, 95, 100, 105, 110, 115, 120, 125,128])
y = np.array([210,240, 250, 260, 270, 280, 290, 300, 310, 320, 330,355])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Sports Watch Data", fontdict = font1,loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.plot(x, y,lw=1.5,c='purple',ls='-',marker='o',ms=8,mec='purple',mfc='r')
plt.show()
```

#### Matplotlib Add Grid Lines to a Plot

##### Specify Which Grid Lines to Display
>You can use the **axis parameter in the grid()** function to **specify which grid lines to display**.
>Legal values are: **'x', 'y', and 'both'. Default value is 'both'**.
##### Set Line Properties for the Grid
>You can also set the line properties of the grid, like this: **grid(color = 'color', linestyle = 'linestyle', linewidth = number)**.

```python
x = np.array([20,80, 85, 90, 95, 100, 105, 110, 115, 120, 125,128])
y = np.array([210,240, 250, 260, 270, 280, 290, 300, 310, 320, 330,355])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Sports Watch Data", fontdict = font1,loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.plot(x, y,lw=1.5,c='purple',ls='-',marker='o',ms=8,mec='purple',mfc='r')
plt.grid(axis='both',c='green',ls='--',lw=1)
plt.show()
```

# Matplotlib Scatter
>With Pyplot, you can use the **scatter() function to draw a scatter plot**.
>The **scatter()** function **plots one dot for each observation**. It needs **two arrays of the same length**, one for the values of the **x-axis**, and one for values on the **y-axis**.

```python
# Creating a simple scatter
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.xlabel("Car Age")
plt.ylabel("Speed of the Car")
plt.scatter(x, y)
plt.show()
```

```python
#Comparing Plots
#In the example above, there seems to be a relationship between speed and age, 
#but what if we plot the observations from another day as well? what  Will the scatter plot tell us ?

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.xlabel("Car Age")
plt.ylabel("Speed of the Car")
plt.show()
```

*By comparing the two plots,we can say that they both gives us the same conclusion,that is newer the car, the faster it drives.*

##### ColorMap

###### Usage of colormap
>Specify the colormap with the keyword argument **cmap** with the value of the colormap.In addition we have to create an array with values (from 0 to 100), one value for each of the point in the scatter plot

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y,c=colors,cmap='viridis')
plt.show()
```

# Pie Chart
>Pie chart visualize **absolute and relative frequencies**.A pie chart is a circe partitioned into segments where each of the segments represents a **category**.The size of the each segment depends upon the relative frequency and it is determied by the angle.A pie chart is a **representation of values as slices of circles with different colors**. 
>>A pie chart is created using **pie()** function considering single list as an argument.

```python
list1=[5,18,43,97,34,67,24,50,92,81]
plt.pie(list1,labels=list1)
plt.show()
```

As you can see the pie chart draws one piece (called a **wedge**) for each value in the list1.By default the plotting of the first wedge starts from the **x-axis** and moves **counterclockwise**.

The size of each wedge is determined by comparing the value with all the other values, by using the formula:
**The value divided by the sum of all values**.i.e., $$\frac{x}{sum(x)}$$


```python
#adding labels by using labels argument
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Mangoes", "Oranges"]

plt.pie(y, labels = mylabels)
plt.show() 
```

##### Start Angle
>As mentioned the default start angle is at the x-axis, but you can **change the start angle by specifying a startangle parameter**.The startangle parameter is defined with an angle in degrees, **default angle is 0**

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Mangoes", "Oranges"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

##### Explode
>when you want one of the wedges to stand out? The explode parameter allows you to do that.
>>The explode parameter, if specified, and not None, must be an array with one value for each wedge.
>>Each value represents how far from the center each wedge is displayed

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Mangoes", "Oranges"]
myexplode = [0.2,0,0,0.05]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

##### Shadow
> Add a shadow to the pie chart by setting the **shadows parameter to True**.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Mangoes", "Oranges"]
myexplode = [0.2,0,0,0]

plt.pie(y, labels=mylabels, explode=myexplode,startangle=90,shadow=True)
plt.show() 
```

##### Colors
>You can set the color of each wedge with the colors parameter.The **colors parameter, if specified, must be an array with one value for each wedge**.

Some of the shortcuts to use colors:
**'r' - Red,'g' - Green,'b' - Blue,'c' - Cyan,'m' - Magenta,'y' - Yellow,'k' - Black,'w' - White**

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Grapes", "Oranges"]
myexplode = [0.05,0.05,0.05,0.05]
mycolors = ["red", "yellow", "#4CAF50","Orange"]
plt.pie(y, labels=mylabels, explode=myexplode,startangle=90,shadow=True,colors=mycolors)
plt.show() 
```

##### Legend
>To add a list of explanation for each wedge, use the **legend()** function

##### Legend With Header
>To add a header to the legend, add the **title parameter to the legend()** function.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Grapes", "Oranges"]
myexplode = [0.05,0.05,0.05,0.05]
mycolors = ["red", "yellow", "#4CAF50","Orange"]
plt.pie(y, labels=mylabels, explode=myexplode,startangle=90,shadow=True,colors=mycolors)
plt.legend(title='Fruits')
plt.show() 
```

# Histograms
>Histogram is based on the idea **to categorise data into different groups and plot the bars for each category with height**.A histogram **represents the frequencies** of values of a variable gathered into ranges.Each bar in a histogram represents the height of the number of values present in the range.(or) A histogram is a graph showing **frequency distributions**.It is a graph showing the number of observations within each given interval.
>>In Matplotlib, we use the **hist()** function to create histograms.The hist() function will use an **array of numbers to create a histogram**, the array is sent into the function as an argument

```python
# A simple histogram
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

```python
x = np.random.normal(100, 10, 250)

plt.hist(x,orientation='horizontal',color='hotpink',density=True)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
 
#Creating dataset
np.random.seed(12345678)
N_points = 15000
n_bins = 30
 
# Creating distribution
x = np.random.randn(N_points)
y = .8 ** x + np.random.randn(15000) + 30
legend = ['distribution']
 
# Creating histogram
fig, axs = plt.subplots(1, 1,
                        figsize =(10, 7),
                        tight_layout = True)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    axs.spines[s].set_visible(False)

# Remove x, y ticks
axs.xaxis.set_ticks_position('none')
axs.yaxis.set_ticks_position('none')
   
# Add padding between axes and labels
axs.xaxis.set_tick_params(pad = 5)
axs.yaxis.set_tick_params(pad = 5)
 
# Add x, y gridlines
axs.grid(b = True, color ='grey',
        linestyle =':', linewidth = 0.5,
        alpha = 1)
 
# Add Text watermark
fig.text(0.9, 0.15, "Divya Reddy's Plot",
         fontsize = 15,
         color ='hotpink',
         ha ='left',
         va ='bottom',
         alpha = 1)
 
# Creating histogram
N, bins, patches = axs.hist(x, bins = n_bins)
 
# Setting color
fracs = ((N**(1 / 5)) / N.max())
norm = colors.Normalize(fracs.min(), fracs.max())
 
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.gist_rainbow(norm(thisfrac))
    thispatch.set_facecolor(color)

# Adding extra features   
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(legend)
plt.title('My Customized Histogram')
 
# Show plot
plt.show()
```

# Area Plot
>Area plots are pretty much similar to the line plot.They are also known as **stack plots**.These plots can be used to track changes over time from 2 or more related groups that make up one whole category.

# Mesh Grid
>The purpose of meshgrid is to create a **rectangular grid out of an array of x values and an array of y values**.
>>A meshgrid is created using **meshgrid()** function.This chart is used for **analysis of large data**.

```python
# Creating large datasets
val1=np.linspace(-9,9,600)
val2=np.linspace(-5,5,500)

# Creating a meshgrid chart
A,B=np.meshgrid(val1,val2)
plt.figure(1,figsize=(10,15))
plt.subplot(131)
plt.imshow(A+B)
plt.subplot(132)
plt.imshow(A-B)
plt.subplot(133)
plt.imshow(A*B)
```

# Contour Plot
> A contour plot is a graphical technique for representing a 3-D surface and is created using **contour()**.It is drawn by plotting constant z slices,called **contour**,on a 2-D format
>>Which means ,given a value "z",lines are drawn  for connecting the (x,y) coordinates where that "z" value occurs.It basically tells us how "x" and "y" variables impact the variable "z".

```python
# Creating large datasets
val1=np.linspace(-9,9,600)
val2=np.linspace(-5,5,500)

# Creating a contour plot
A,B=np.meshgrid(val1,val2)
plt.figure(1,figsize=(15,10))
plt.subplot(2,4,1)
plt.contour(A,B,A+B,alpha=0.75,cmap='jet')
plt.subplot(2,4,2)
plt.contourf(A,B,A+B,alpha=0.75,cmap='jet')
plt.subplot(2,4,3)
plt.contour(A,B,A*B,alpha=0.75,cmap='jet')
plt.subplot(2,4,4)
plt.contourf(A,B,A*B,alpha=0.75,cmap='jet')
plt.subplot(2,4,5)
plt.contour(A,B,A-B,alpha=0.75,cmap='jet')
plt.subplot(2,4,6)
plt.contourf(A,B,A-B,alpha=0.75,cmap='jet')
plt.subplot(2,4,7)
plt.contour(A,B,B-A,alpha=0.75,cmap='jet')
plt.subplot(2,4,8)
plt.contourf(A,B,B-A,alpha=0.75,cmap='jet')


plt.suptitle('Contour Plot')
plt.show()
```

```python
# invent some numbers, turning the x and y arrays into simple2d arrays, which make combining them together easier.
x = np.linspace(-3, 15, 50).reshape(1, -1)
y = np.linspace(-3, 15, 20).reshape(-1, 1)
z = np.cos(x)*2 - np.sin(y)*2
  
# we no longer need x and y tobe 2 dimensional, so flatten them.
x, y = x.flatten(),y.flatten()
cs = plt.contourf(x, y, z, hatches =['-', '/','\\', '//'],cmap ='gist_rainbow',extend ='both',alpha = 1)
plt.colorbar(cs)
  
plt.title('Contourf Plot')
plt.show()
```
